```python
import csv
import math
import urllib.parse

import matplotlib.colors
import matplotlib.pyplot as plt
import numpy
import scipy.cluster.hierarchy as hcluster
```

```python
FILENAMES = [
    'last1.tsv.txt',
    'last2.tsv.txt',
]
```

```python
dimensions_count = None
names_table = []
hidden_table = []
preloaded_data_table = []

for filename in FILENAMES:
    with open(filename) as file:
        csvreader = csv.reader(file, delimiter='\t')
        preloaded_data = list(csvreader)
    if dimensions_count is None:
        dimensions_count = len(preloaded_data[0])
    assert dimensions_count == len(preloaded_data[0])
    for line in preloaded_data[1:]:  # skipping headers
        names_table.append(line[0])
        dataline = list(map(int, line[1:1+6]))
        preloaded_data_table.append(dataline)
        processed_flag = len(line[7]) > 0
        hidden_table.append(processed_flag)

data_table = numpy.array(preloaded_data_table)
```

MAXCLUST_THRESHOLD = 11
linkage_matrix = hcluster.linkage(data_table, metric='cityblock', method='average')
clusters_table = hcluster.fcluster(linkage_matrix, MAXCLUST_THRESHOLD, criterion='maxclust')
clusters_count = max(clusters_table)
print(repr(clusters_count))

```python
DISTANCE_THRESHOLD = 3.8
linkage_matrix = hcluster.linkage(data_table, metric='cityblock', method='average')
clusters_table = hcluster.fcluster(linkage_matrix, DISTANCE_THRESHOLD, criterion='distance')
clusters_count = max(clusters_table)
print(repr(clusters_count))
```

```python
result_table = list(zip(names_table, preloaded_data_table, hidden_table, list(clusters_table)))

grouped_results = [[] for i in range(clusters_count)]
for r in result_table:
    grouped_results[r[3]-1].append((r[0], r[1], r[2]))
```

```python
DENDROGRAM_TRUNCATE_VAL = 64
dendrogram_figsize = (DENDROGRAM_TRUNCATE_VAL*0.3, DENDROGRAM_TRUNCATE_VAL*0.2)
dendrogram_labels = ["{}. {}".format(t[3], urllib.parse.unquote(t[0])) for t in result_table]
plt.figure(figsize=dendrogram_figsize)
hcluster.dendrogram(linkage_matrix, p=DENDROGRAM_TRUNCATE_VAL, truncate_mode='lastp', orientation='left', labels=dendrogram_labels, show_leaf_counts=True)
plt.show()
```

```python
def _get_color(nval, param_i, val_i):
    if param_i <= 1:
        hue = (110, 60, 30, 0)[val_i] / 360
    else:
        hue = (110, 60, 60, 45)[val_i] / 360
    saturation = nval
    value = 1
    return matplotlib.colors.hsv_to_rgb((hue, saturation, value))

MULTIPLOT_COLUMNS_COUNT = 3
multiplot_rows_count = int(math.ceil(clusters_count/MULTIPLOT_COLUMNS_COUNT))
multiplot_dim_sizes = (multiplot_rows_count, MULTIPLOT_COLUMNS_COUNT)
multiplot_figsize = (multiplot_dim_sizes[1]*5, multiplot_dim_sizes[0]*4)
plt.figure(figsize=multiplot_figsize)
for cluster_i in range(clusters_count):
    data_points_matrix = numpy.array([t[1] for t in grouped_results[cluster_i]])
    dimension_histograms_listy = [
        numpy.histogram(data_points_matrix[:, col_i], bins=4, range=(0., 4.))[0]
        for col_i in range(data_points_matrix.shape[1])]
    histogram_maxs = list(map(max, dimension_histograms_listy))
    colormap_listy = [
        [_get_color(dimension_histograms_listy[param_i][val_i]/histogram_maxs[param_i], param_i, val_i) for val_i in range(4)]
        for param_i in range(6)]
    colormap_matrix = numpy.array(colormap_listy)
    axes = plt.subplot(*multiplot_dim_sizes, cluster_i+1)
    axes.set_xticks(numpy.arange(6), ['A1', 'A2', 'B1', 'B2', 'B3', 'B4'])
    axes.set_yticks(numpy.arange(5), [str(int(x)) for x in numpy.linspace(0, 4, 5)])
    axes.set_title("Cluster #{} (count: {})".format(cluster_i+1, data_points_matrix.shape[0]))
    axes.imshow(colormap_matrix.transpose((1, 0, 2)), origin='lower')
    for param_i in numpy.arange(6):
        for val_i in numpy.arange(4):
            val = dimension_histograms_listy[param_i][val_i]
            if val > 0:
                axes.text(param_i, val_i, val, ha='center', va='center', color='black')
plt.show()
```
